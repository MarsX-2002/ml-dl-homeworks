```python
# data
names = open("names.txt").read().splitlines()

names
```

```python
# 3-gram counts
counts = {}

for name in names:
    name = ["<s>", "<s>"] + list(name) + ["</s>"]  # two start tokens in 3-gram

    for i in range(len(name) - 2):
        prefix = (name[i], name[i + 1])  # first two chars of trigram
        next_char = name[i + 2]  # third char of trigram

        if prefix not in counts:
            counts[prefix] = {}

        if next_char not in counts[prefix]:
            counts[prefix][next_char] = 0

            # print("pref", counts[prefix]) 
            # print("counts",counts[prefix][next_char])   
        counts[prefix][next_char] += 1 


counts        

```

```python
# convert to probs
probs = {}

for prefix, next_chars in counts.items():
    total = sum(next_chars.values())
    probs[prefix] = {}  # inner dict
    # print(f"({ch1} , {ch2}), {next_char} == > {total}\n")
    for char, count in next_chars.items():
        probs[prefix][char] = count / total

probs

```

```python
# generate names
import numpy as np

def generate_names():
    name = ['<s>', '<s>']

    while True:
        prefix = tuple(name[-2:])  # last 2 chars

        if prefix not in probs:
            break

        # pick next char based on probabilities
        chars = list(probs[prefix].keys())
        weights = np.array(list(probs[prefix].values()))
        next_char = chars[np.argmax(np.random.multinomial(1, weights))]

        if next_char == '</s>':
            break

        name.append(next_char)

    return ''.join(name[2:])


for _ in range(10):
    print(generate_names())    

        
    
        
```
